<p style="text-align: justify;">4-Construya un Panel con los tres (3) dataframes cargados anteriormente: df1, df2 y df3. Para construir este panel, cree un diccionario llamado panel que en sus keys tenga los nombres de los conjuntos de datos y los valores sean los dataframes.</p>

```python
#construir un Panel con los tres DataFrames
panel = {'df1': df1, 'df2': df2, 'df3': df3}# creación del diccionario con sus llaves, donde están los nombres de los 
#anteriores dataframes cargados.
```

<p style="text-align: justify;">5-Programe una función que extraiga todas las listas de los encabezados de columnas que se encuentran en un Panel yúsela para extraer las listas de columnas del Panel del punto anterior. Imprima los resultados de forma que se diferencien los encabezados entre los dataframes en el Panel.</p>

Se procede con crear las respectivas funciones python para extraer las listas de los encabezados, según analizado en el material de Grus (2023):

```python
"""
En el siguiente codigo, se define una función llamada extraer_encabezados de los 3
dataframes, cargados en las variables df1 df2 df3
"""

def extraer_encabezados(panel):# se define la función extraer_encabezados de todos los dataframes
    encabezados = {}
    for nombre, df in panel.items():# ciclo for que va a tomar todos titulos de las columnas de los dataframes 
        encabezados[nombre] = df.columns.tolist()# impresión de los datos en una lista
    return encabezados #devolver valores especificos de donde se llamó la función.


encabezados = extraer_encabezados(panel)# se crea una variable llamada encabezados, que invoque la función extraer_encabezados
print("Encabezados de las columnas en el Panel:")# impresión de mensaje
for nombre, cols in encabezados.items():# ciclo for para que proceda a imprimir los titulos de los 3 dataframes.
    print(f"{nombre}: {cols}") # impresión de mensaje
```

<p style="text-align: justify;">6-Programe una función que imprima la cantidad de elementos que tienen las bases de datos en un Panel y úsela con el Panel del punto anterior. Imprima los resultados de forma que se diferencien los tamaños de los diferentes dataframes en el Panel.</p>

Se realliza el procedimiento conforme a lo consultado en el matierial de Grus (2023), que es crear funciones para imprimir las dimensiones de los DataFrame.

```python
# Paso 5: Programar una función para imprimir la cantidad de elementos en cada DataFrame del Panel
def imprimir_tamanos(panel):# se define función para imprimit los tamaños o dimensiones del dataframe
    tamanos = {}
    for nombre, df in panel.items():# ciclo for para que tome los valores del panel y después la función shape para obtener los numeros de filas y columnas de los df
        tamanos[nombre] = df.shape
    return tamanos#devolver valores especificos de donde se llamó la función.

tamanos = imprimir_tamanos(panel)
print("Cantidad de elementos en cada DataFrame del Panel:")# impresión de mensaje
for nombre, tamano in tamanos.items():# ciclo for para que proceda a imprimir los tamaños de los conjuntos de datos de todo el panel
    print(f"{nombre}: {tamano[0]} filas, {tamano[1]} columnas")
```

7-Filtre el dataframe df1 para obtener las muestras de la empresa TAC. Determine cuantos elementos tiene dicho filtro.

```python
df1_filtrado = df1[df1['Empresa'] == 'TAC']# se utiliza una xxx tomar del dataframe la columna "Empresa" que filtre la palabra TAC

print("El DataFrame filtrado, en la columna 'EMPRESA' tiene muestras TAC:")# impresión de mensaje
print(f"Número de filas: {df1_filtrado.shape[0]}")# impresión de mensaje donde muestra la cantidad de filas en el df
print(f"Número de columnas: {df1_filtrado.shape[1]}")# impresión de mensaje donde muestra la cantidad de columnas en el df
```

8-Determine cuantos elementos nulos hay en cada una de las columnas del dataframe df2.

```python
#Determinar cuántos elementos nulos hay en cada columna de df2
valores_nulos_por_columna = df2.isnull().sum()# se define variable donde se convoca una función llamada isnull() y .sum() para
#que detecte y sume los valores nulos en el df2.

print("Número de elementos nulos por columna en el df2:")# impresión de mensaje
print(valores_nulos_por_columna)# impresión de mensaje

suma_total_valores_nulos =valores_nulos_por_columna.sum()# sumtaoria de los valores


print('La sumatoria de los elementos en el DataFrame 2 (df2) es de: ',suma_total_valores_nulos, 'valores nulos en todas las filas y columnas')
```
