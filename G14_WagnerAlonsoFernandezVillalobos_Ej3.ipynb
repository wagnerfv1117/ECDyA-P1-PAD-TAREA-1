<h3><left>2-Mostrar la funci√≥n lineal ùíá(ùíô) = ùüèùüïùíô ‚àí ùüêùüèùüé. ùüëùüëùüëùüë en el dominio
[13,18].<h3>

Desarrollo del c√≥digo:

* Se generar√° un gr√°fico de la funci√≥n lineal ùíá(ùíô) = ùüèùüïùíô ‚àí ùüêùüèùüé. ùüëùüëùüëùüë desde el dominio 13 al 18.

```python
x = np.linspace(13, 18, 100)#se procede con generar 100 puntos en el dominio [desde 13 a 18]
y = 17*x - 210.3333#se calculan los valores correspondientes a la funci√≥n lineal

plt.plot(x, y, '--')#se utiliza el estilo de l√≠nea punteada como lo pide la guia de aprendizaje

"""Las siguientes lineas son para graficar la funci√≥n lineal"""
plt.xlabel('eje x')
plt.ylabel('eje y = f(x)')
plt.title('Gr√°fico No. 2 - Funci√≥n lineal')
plt.grid(True)
plt.show()


x = np.linspace(13, 18, 100)#funci√≥n que devuelve n√∫meros espaciados uniformemente durante un intervalo especificado.
y = 17*x - 210.3333#se alcula los valores correspondientes a la funci√≥n lineal, es decir ùíá(ùíô) = ùüèùüïùíô ‚àí ùüêùüèùüé. ùüëùüëùüëùüë

plt.plot(x, y, '--')  # funci√≥n para utilizar el estilo de l√≠nea punteada

"""Agregar puntos de dispersi√≥n, complementandolo en la grafica lineal
   se crean las listas y la ecuaci√≥n ùíá(ùíô) = ùüèùüïùíô ‚àí ùüêùüèùüé. ùüëùüëùüëùüë
"""
x_puntos = [13, 14, 15, 16, 17, 18]
y_puntos = [17*x - 210.3333 for x in x_puntos]
plt.scatter(x_puntos, y_puntos, color='orange')#color a los puntos de dspersi√≥n


"""desarrollo del codigo para plotear la funci√≥n
"""
plt.xlabel('eje x')# asignar titulo al eje x
plt.ylabel('eje y = f(x)')#asignar titulo al eje y
plt.title('Gr√°fico No. 3 - Funci√≥n lineal con puntos de dispersi√≥n')
plt.grid(True)
plt.show()
```

<h3><left>3-Mostrar la funci√≥n exponencial ùíà(ùíô) = ùüé. ùüèùüíùüìùüê ùíÜ(ùüé.ùüëùüîùüóùíô) en el dominio [13,18]<h3>

Desarrollo del c√≥digo:

* Se crear√° el gr√°fico de la funci√≥n exponencial ùíà(ùíô) = ùüé. ùüèùüíùüìùüê ùíÜ^(ùüé.ùüëùüîùüóùíô) en el dominio [13, 18] con l√≠neas punteadas, puedes utilizar la biblioteca matplotlib en Python.

* Se generar√° el gr√°fico de la funci√≥n exponencial ùíà(ùíô) = ùüé. ùüèùüíùüìùüê ùíÜ^(ùüé.ùüëùüîùüóùíô) en el dominio [13, 18] con l√≠neas punteadas. 

* Se invoca la funci√≥n linspace, para generar 100 puntos equidistantes en el dominio [13, 18]. 

* Se calcula los valores correspondientes a la funci√≥n exponencial utilizando la ecuaci√≥n dada. muestra el gr√°fico.

* Se generar√° el gr√°fico de la funci√≥n exponencial ùíà(ùíô) = ùüé. ùüèùüíùüìùüê ùíÜ^(ùüé.ùüëùüîùüóùíô) en el dominio [13, 18] con l√≠neas punteadas y puntos de dispersi√≥n.

```python
x = np.linspace(13, 18, 100)#Se generan 100 puntos en el dominio [13, 18]
y = 0.1452 * np.exp(0.369*x)#Se calculan los valores correspondientes a la funci√≥n exponencial

plt.plot(x, y, '--')# se utiliza el estilo de l√≠nea punteada.

#codigos para a√±adir las leyenda del grafico
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('Gr√°fico No. 4 - Funci√≥n exponencial')
plt.grid(True)
plt.show()


x = np.linspace(13, 18, 100)  #Generaci√≥n de los 100 puntos en el dominio [13, 18]
y = 0.1452 * np.exp(0.369*x)  #Calculo de los valores correspondientes a la funci√≥n exponencial

plt.plot(x, y, '--')  #estilo de l√≠nea punteada

#se agregan los puntos de dispersi√≥n
x_puntos = [13, 14, 15, 16, 17, 18]
y_puntos = [0.1452 * np.exp(0.369*x) for x in x_puntos]#aplicaci√≥n de la exponencial en el grafico.
plt.scatter(x_puntos, y_puntos, color='green')# color verde los puntos


#graf√≠ca con los puntos de dispersi√≥n
plt.xlabel('x')
plt.ylabel('eje y = g(x)')
plt.title('Gr√°fico No. 5 - Funci√≥n exponencial con puntos de dispersi√≥n')
plt.grid(True)
plt.show()
```

<h3><left>4-Mostrar la funci√≥n logar√≠tmica ùíâ(ùíô) = ùüêùüìùüó. ùüíùüó ùíçùíè(ùíô) ‚àí ùüîùüìùüî. ùüíùüî en el
dominio [13,18].<h3>

Desarrollo del c√≥digo:

* Se elabora la funci√≥n logar√≠tmica ùíâ(ùíô) = ùüêùüìùüó. ùüíùüó ùíçùíè(ùíô) ‚àí ùüîùüìùüî. ùüíùüî en el dominio [13, 18] con l√≠neas punteadas, usando la biblioteca matplotlib en Python.
* Se genera la funci√≥n logar√≠tmica ùíâ(ùíô) = ùüêùüìùüó. ùüíùüó ùíçùíè(ùíô) ‚àí ùüîùüìùüî. ùüíùüî desde el dominio 13 al 18 con l√≠neas punteadas, La funci√≥n linspace se utiliza para generar 100 ,para calcular los valores correspondientes a la funci√≥n logar√≠tmica utilizando la ecuaci√≥n.

```python
x = np.linspace(13, 18, 100)  # Generar 100 puntos en el dominio [13, 18]
y = 259.49 * np.log(x) - 566.46  # Calcular los valores correspondientes a la funci√≥n logar√≠tmica

plt.plot(x, y, '--')#utilizar el estilo de l√≠nea punteada
plt.xlabel('x')
plt.ylabel('h(x)')
plt.title('Gr√°fico No. 6 - funci√≥n logar√≠tmica')
plt.grid(True)
plt.show()


x = np.linspace(13, 18, 100)  #generar 100 puntos en el dominio [13, 18]
y = 259.49 * np.log(x) - 566.46  #calcular los valores correspondientes a la funci√≥n logar√≠tmica

plt.plot(x, y, '--')  #utilizar el estilo de l√≠nea punteada

#agregar puntos de dispersi√≥n
x_puntos = [13, 14, 15, 16, 17, 18]
y_puntos = [259.49 * np.log(x) - 566.46 for x in x_puntos]
plt.scatter(x_puntos, y_puntos, color='red')

plt.xlabel('Eje x')
plt.ylabel('Eje y = h(x)')
plt.title('Gr√°fico No. 7 - Funci√≥n logar√≠tmica con puntos de dispersi√≥n')
plt.grid(True)
plt.show()
```

<h3><left>Generar 4 listas aleatorias de 700 muestras (revise la
documentaci√≥n de numpy.random):


-x1 con distribuci√≥n normal de media 5 y desviaci√≥n
est√°ndar 0.1.

-y1 con distribuci√≥n normal de media 7 y desviaci√≥n
est√°ndar 0.1.

-x2 con distribuci√≥n uniforme entre el valor m√≠nimo de x1 y
el valor m√°ximo de x1.

-y2 con distribuci√≥n uniforme entre el valor m√≠nimo de y1 y
el valor m√°ximo de y1.

-Graficar en un mismo plot nubes de dispersi√≥n de los pares de
puntos que se forman con (x1, y1) y (x2, y2).

-La nube de puntos (x2, y2) debe visualizarse sobre la nube de
(x1, y1).

-Anexe los t√≠tulos y legendas necesarios para poder citar y

Desarrollo del c√≥digo:

* Creaci√≥n de lista aleatorias para las 700 muestras utilizando la librer√≠a Numpy con los modulos np.random.
* Creaci√≥n de muestra con valores minimos y maximos para una distribuci√≥n uniforme de los datos.
* Grafico de las funciones anteriormente descritas.
* Creaci√≥n de titulos y leyendas.


```python
#generaci√≥n de las listas aleatorias con sus respectivos valores minimos y maximos.
x1 = np.random.normal(5, 0.1, 700)#700 muestras
y1 = np.random.normal(7, 0.1, 700)#700 muestras
x2 = np.random.uniform(np.min(x1), np.max(x1), 700)# VALORES MINIMOS PARA UNA DISTRIBUCION UNIFORME
y2 = np.random.uniform(np.min(y1), np.max(y1), 700)# VALORES MAXIMOS PARA UNA DISTRIBUCI√ìN UNIFORME

#Se procede con graficar las nubes de dispersi√≥n
plt.scatter(x1, y1, label='Nube de dispersi√≥n 1')
plt.scatter(x2, y2, label='Nube de dispersi√≥n 2')

#titulos y graficos y leyendas 
plt.title('Gr√°fico No. 8 - Nubes de dispersi√≥n')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend()

#Plotear el grafico de dispersi√≥n
plt.show()
```

<h3><left>Cargar los datos del archivo Excel "Anexo 1 - Base de Datos Excel.xlsx" en la variable df1 y generar los gr√°ficos solicitados, puedes utilizar la biblioteca pandas y matplotlib en Python. <h2>

Desarrollo del c√≥digo:

* Se utiliza funci√≥n pd.read_excel() de pandas para cargar los datos del archivo Excel  en la variable df1.
* Se acomoda los nombres de las columnas del DataFrame, omitiendo las 3 primeras filas que tiene el archivo Excel (.xlsx) 
* Se utiliza la funci√≥n plt.hist() de matplotlib para generar los histogramas de las variables "Tiempo motor ON" y "Tiempo en Movimiento". 
* Se utiliza el par√°metro alpha para hacer los histogramas lo suficientemente transparentes y se puedan visualizar ambos gr√°ficos. 
* Se agrega t√≠tulos, leyendas y etiquetas a los ejes para facilitar la interpretaci√≥n de los histogramas.
* Se utiliza la funci√≥n plt.scatter() de Matplotlib, para generar el gr√°fico de dispersi√≥n emparejando los valores de "Tiempo motor ON" y "Tiempo en Movimiento".

```python
df1=pd.read_excel('data/Anexo 1 - Base de Datos EXCEL.xlsx', sheet_name='BASE DE DATOS', skiprows=3)

#proceso para justar el nombre de las columnas.
asignar_nombres_columnas = ['Servicio', 'Empresa', 'Turno', 'D√≠a de la semana', 'Experiencia del conductor',
                    'Formaci√≥n del Param√©dico', 'Nivel del Param√©dico', 'Cantidad de atenciones',
                    'Distancia recorrida (Km)', 'Velocidad m√°xima (Km/h)', 'Consumo de combustible (Galones)',
                    'Tiempo en Movimiento (h)', 'Tiempo motor ON (h)'] 

df1.columns = asignar_nombres_columnas# funci√≥n para asignar nuevamente el nombre de las columnas


#se procede con generar los histogramas
plt.hist(df1['Tiempo motor ON (h)'], alpha=0.5, label='Tiempo motor ON - (Horas)', density=True)#tiempo motor on
plt.hist(df1['Tiempo en Movimiento (h)'], alpha=0.5, label='Tiempo en Movimiento - (Horas)', density=True)#tiempo del motor en movimiento

#se agregan los titulos y leyendas al histograma
plt.title('Gr√°fico No. 8 - Histogramas de Tiempo motor ON y Tiempo en Movimiento')
plt.xlabel('Tiempo (Horas)')
plt.ylabel('Frecuencia o conteo')
plt.legend()
plt.show()

#se procede con generar el grafico de dispersi√≥n
plt.scatter(df1['Tiempo motor ON (h)'], df1['Tiempo en Movimiento (h)'])

#agregar t√≠tulos y leyendas
plt.title('Gr√°fico No. 9 - Gr√°fico de dispersi√≥n de Tiempo motor ON y Tiempo en Movimiento')
plt.xlabel('Tiempo motor ON - (Horas)')
plt.ylabel('Frecuencia o conteo')
plt.show()
```

<h4><left>PUNTO No. 3.1.HISTOGRAMA CON ASMITER√çA</h4>

Se desarrollan los graficos usando la librer√≠a seaborn, para verificar si son simetricos o asim√©tricos, que seg√∫n .Bruce et¬†al. (2022, pp. 24-25), son medidas de distribuci√≥n estadistica, que indica la localizaci√≥n y la variabilidad de los datos en el plano cartesiano; cuando el grafico es uniforme, se considera sim√©trico, cuando el grafico tiene desviaci√≥n hacia la derecha o izquierda se considera asim√©trico, teniendo en cuenta la **media** en las visualizaci√≥nes. 

La media segun Romero & Z√∫nica (2020), son datos estadisticos que hacen parte de las medidas de tendencia central, es la cantidad que se obtiene al sumar todos los datos de un conjunto de valores, para posteriormente dividir la cifra obtenida entre la cantidad de valores analizados. Es de anotar que dichos gr√°ficos cuando tienen datos su parte m√°s baja, pero son largos, es considerado este concepto como _"cola"_ ; si la cola se ubica hacia la izquierda es conocida como asmiter√≠a negativa, si la cola se ubica  hacia la derecha, se denomina asimetr√≠a positiva.

**Ilustraci√≥n No. 1** - Medidas de Distribuci√≥n estad√≠stica

_Fuente_. Extra√≠do de Romero & Z√∫nica (2020)

**Ilustraci√≥n No. 2** - Tipos de simetr√≠a

_Fuente_. Extra√≠do de https://estadisticadescriptiva.com/datos/medidas-de-asimetria/

```python
#crear el gr√°fico de distribuci√≥n utilizando sns.distplot() 
sns.distplot(df1['Tiempo motor ON (h)']) 
sns.distplot(df1['Tiempo en Movimiento (h)']) 
 
#calculo la media de los datos 
media = df1['Tiempo motor ON (h)'].mean() 
media = df1['Tiempo en Movimiento (h)'].mean() 

plt.title('Gr√°fico No. 10 - Histogramas de Tiempo motor ON y Tiempo en Movimiento con curvas de asimetr√≠a')
plt.xlabel("Tiempo en Movimiento (Horas)")
plt.ylabel('Frecuencia')
plt.show()
```

* C√°lculo especifico o detallado de la  media en la columna Tiempo en movimiento

```python
plt.figure(figsize=(10, 6))# configuraci√≥n del tama√±o del gr√°fico


plt.hist(df1['Tiempo motor ON (h)'], 
         bins=20, color='blue', 
         alpha=0.7, 
         label='Tiempo motor ON (h)')#para crear grafico de histograma de la columna Tiempo motor ON


media_tiempo_motor = df1['Tiempo motor ON (h)'].mean()#media de Tiempo motor ON

plt.axvline(x=media_tiempo_motor, 
            color='red', 
            linestyle='--', 
            label=f'Media: {media_tiempo_motor:.2f} h')# crear la linea de la media


plt.hist(df1['Tiempo en Movimiento (h)'], 
         bins=20, 
         color='orange', 
         alpha=0.7, label='Tiempo en Movimiento (h)')# para crear la grafica de la columna Tiempo en movimiento


media_tiempo_movimiento = df1['Tiempo en Movimiento (h)'].mean()#media de Tiempo en Movimiento (h)

plt.axvline(x=media_tiempo_movimiento, 
            color='red', 
            linestyle='--', 
            label=f'Media: {media_tiempo_movimiento:.2f} h')# crear la linea de la media


plt.xlabel('Tiempo (Horas)')
plt.ylabel('Frecuencia')
plt.title('Gr√°fico No. 11 - Histogramas Tiempo motor ON y Tiempo en Movimiento (con lineas de media)')
plt.legend()#se a√±ade leyenda a√± gr√°fico para que mueste el dato de la media."
plt.show()
```
