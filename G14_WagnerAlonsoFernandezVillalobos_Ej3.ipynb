<h3><left>1- Mostrar los puntos de dispersión del emparejamiento las

<h3><left>2-Mostrar la función lineal 𝒇(𝒙) = 𝟏𝟕𝒙 − 𝟐𝟏𝟎. 𝟑𝟑𝟑𝟑 en el dominio
[13,18].<h3>

Desarrollo del código:

* Se generará un gráfico de la función lineal 𝒇(𝒙) = 𝟏𝟕𝒙 − 𝟐𝟏𝟎. 𝟑𝟑𝟑𝟑 desde el dominio 13 al 18. 
* Se plotea en grafico en el plano cartesiano, con lineas punteadas, usando la función plt.plot().
* Codigos para añadir leyendas al grafico, usando la función _plt.label()_.
* Código para añadirlos los puntos de dispersión o scatter del 13 al 18.


```python
x = np.linspace(13, 18, 100)#se procede con generar 100 puntos en el dominio [desde 13 a 18]
y = 17*x - 210.3333#se calculan los valores correspondientes a la función lineal

plt.plot(x, y, '--')#se utiliza el estilo de línea punteada como lo pide la guia de aprendizaje

"""Las siguientes lineas son para graficar la función lineal"""
plt.xlabel('eje x')
plt.ylabel('eje y = f(x)')
plt.title('Gráfico No. 2 - Función lineal')
plt.grid(True)
plt.show()


x = np.linspace(13, 18, 100)#función que devuelve números espaciados uniformemente durante un intervalo especificado.
y = 17*x - 210.3333#se alcula los valores correspondientes a la función lineal, es decir 𝒇(𝒙) = 𝟏𝟕𝒙 − 𝟐𝟏𝟎. 𝟑𝟑𝟑𝟑

plt.plot(x, y, '--')  # función para utilizar el estilo de línea punteada

"""Agregar puntos de dispersión, complementandolo en la grafica lineal
   se crean las listas y la ecuación 𝒇(𝒙) = 𝟏𝟕𝒙 − 𝟐𝟏𝟎. 𝟑𝟑𝟑𝟑
"""
x_puntos = [13, 14, 15, 16, 17, 18]
y_puntos = [17*x - 210.3333 for x in x_puntos]
plt.scatter(x_puntos, y_puntos, color='orange')#color a los puntos de dspersión


"""desarrollo del codigo para plotear la función
"""
plt.xlabel('eje x')# asignar titulo al eje x
plt.ylabel('eje y = f(x)')#asignar titulo al eje y
plt.title('Gráfico No. 3 - Función lineal con puntos de dispersión')
plt.grid(True)
plt.show()
```

<h3><left>3-Mostrar la función exponencial 𝒈(𝒙) = 𝟎. 𝟏𝟒𝟓𝟐 𝒆(𝟎.𝟑𝟔𝟗𝒙) en el dominio [13,18]<h3>

Desarrollo del código:

* Se creará el gráfico de la función exponencial 𝒈(𝒙) = 𝟎. 𝟏𝟒𝟓𝟐 𝒆^(𝟎.𝟑𝟔𝟗𝒙) en el dominio [13, 18] con líneas punteadas, puedes utilizar la biblioteca matplotlib en Python.

* Se generará el gráfico de la función exponencial 𝒈(𝒙) = 𝟎. 𝟏𝟒𝟓𝟐 𝒆^(𝟎.𝟑𝟔𝟗𝒙) en el dominio [13, 18] con líneas punteadas. 

* Se invoca la función linspace, para generar 100 puntos equidistantes en el dominio [13, 18]. 

* Se calcula los valores correspondientes a la función exponencial utilizando la ecuación dada. muestra el gráfico.

* Se generará el gráfico de la función exponencial 𝒈(𝒙) = 𝟎. 𝟏𝟒𝟓𝟐 𝒆^(𝟎.𝟑𝟔𝟗𝒙) en el dominio [13, 18] con líneas punteadas y puntos de dispersión.

```python
x = np.linspace(13, 18, 100)#Se generan 100 puntos en el dominio [13, 18]
y = 0.1452 * np.exp(0.369*x)#Se calculan los valores correspondientes a la función exponencial

plt.plot(x, y, '--')# se utiliza el estilo de línea punteada.

#codigos para añadir las leyenda del grafico
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('Gráfico No. 4 - Función exponencial')
plt.grid(True)
plt.show()


x = np.linspace(13, 18, 100)  #Generación de los 100 puntos en el dominio [13, 18]
y = 0.1452 * np.exp(0.369*x)  #Calculo de los valores correspondientes a la función exponencial

plt.plot(x, y, '--')  #estilo de línea punteada

#se agregan los puntos de dispersión
x_puntos = [13, 14, 15, 16, 17, 18]
y_puntos = [0.1452 * np.exp(0.369*x) for x in x_puntos]#aplicación de la exponencial en el grafico.
plt.scatter(x_puntos, y_puntos, color='green')# color verde los puntos


#grafíca con los puntos de dispersión
plt.xlabel('x')
plt.ylabel('eje y = g(x)')
plt.title('Gráfico No. 5 - Función exponencial con puntos de dispersión')
plt.grid(True)
plt.show()
```

<h3><left>Generar 4 listas aleatorias de 700 muestras (revise la
documentación de numpy.random):

estándar 0.1.

-x2 con distribución uniforme entre el valor mínimo de x1 y

-y2 con distribución uniforme entre el valor mínimo de y1 y

-Graficar en un mismo plot nubes de dispersión de los pares de
puntos que se forman con (x1, y1) y (x2, y2).

-La nube de puntos (x2, y2) debe visualizarse sobre la nube de

-Anexe los títulos y legendas necesarios para poder citar y

Desarrollo del código:

* Creación de lista aleatorias para las 700 muestras utilizando la librería Numpy con los modulos np.random.
* Creación de muestra con valores minimos y maximos para una distribución uniforme de los datos.
* Grafico de las funciones anteriormente descritas.
* Creación de titulos y leyendas.


```python
#generación de las listas aleatorias con sus respectivos valores minimos y maximos.
x1 = np.random.normal(5, 0.1, 700)#700 muestras
y1 = np.random.normal(7, 0.1, 700)#700 muestras
x2 = np.random.uniform(np.min(x1), np.max(x1), 700)# VALORES MINIMOS PARA UNA DISTRIBUCION UNIFORME
y2 = np.random.uniform(np.min(y1), np.max(y1), 700)# VALORES MAXIMOS PARA UNA DISTRIBUCIÓN UNIFORME

#Se procede con graficar las nubes de dispersión
plt.scatter(x1, y1, label='Nube de dispersión 1')
plt.scatter(x2, y2, label='Nube de dispersión 2')

#titulos y graficos y leyendas 
plt.title('Gráfico No. 8 - Nubes de dispersión')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend()

#Plotear el grafico de dispersión
plt.show()
```

* Se utiliza la función plt.hist() de matplotlib para generar los histogramas de las variables "Tiempo motor ON" y "Tiempo en Movimiento". 
* Se utiliza el parámetro alpha para hacer los histogramas lo suficientemente transparentes y se puedan visualizar ambos gráficos. 
* Se agrega títulos, leyendas y etiquetas a los ejes para facilitar la interpretación de los histogramas.
* Se utiliza la función plt.scatter() de Matplotlib, para generar el gráfico de dispersión emparejando los valores de "Tiempo motor ON" y "Tiempo en Movimiento".

```python
df1=pd.read_excel('data/Anexo 1 - Base de Datos EXCEL.xlsx', sheet_name='BASE DE DATOS', skiprows=3)

#proceso para justar el nombre de las columnas.
asignar_nombres_columnas = ['Servicio', 'Empresa', 'Turno', 'Día de la semana', 'Experiencia del conductor',
                    'Formación del Paramédico', 'Nivel del Paramédico', 'Cantidad de atenciones',
                    'Distancia recorrida (Km)', 'Velocidad máxima (Km/h)', 'Consumo de combustible (Galones)',
                    'Tiempo en Movimiento (h)', 'Tiempo motor ON (h)'] 

df1.columns = asignar_nombres_columnas# función para asignar nuevamente el nombre de las columnas


#se procede con generar los histogramas
plt.hist(df1['Tiempo motor ON (h)'], alpha=0.5, label='Tiempo motor ON - (Horas)', density=True)#tiempo motor on
plt.hist(df1['Tiempo en Movimiento (h)'], alpha=0.5, label='Tiempo en Movimiento - (Horas)', density=True)#tiempo del motor en movimiento

#se agregan los titulos y leyendas al histograma
plt.title('Gráfico No. 8 - Histogramas de Tiempo motor ON y Tiempo en Movimiento')
plt.xlabel('Tiempo (Horas)')
plt.ylabel('Frecuencia o conteo')
plt.legend()
plt.show()

#se procede con generar el grafico de dispersión
plt.scatter(df1['Tiempo motor ON (h)'], df1['Tiempo en Movimiento (h)'])

#agregar títulos y leyendas
plt.title('Gráfico No. 9 - Gráfico de dispersión de Tiempo motor ON y Tiempo en Movimiento')
plt.xlabel('Tiempo motor ON - (Horas)')
plt.ylabel('Frecuencia o conteo')
plt.show()
```

<h4><left>PUNTO No. 3.1.HISTOGRAMA CON ASMITERÍA</h4>

Se desarrollan los graficos usando la librería seaborn, para verificar si son simetricos o asimétricos, que según .Bruce et al. (2022, pp. 24-25), son medidas de distribución estadistica, que indica la localización y la variabilidad de los datos en el plano cartesiano; cuando el grafico es uniforme, se considera simétrico, cuando el grafico tiene desviación hacia la derecha o izquierda se considera asimétrico, teniendo en cuenta la **media** en las visualizaciónes. 

La media segun Romero & Zúnica (2020), son datos estadisticos que hacen parte de las medidas de tendencia central, es la cantidad que se obtiene al sumar todos los datos de un conjunto de valores, para posteriormente dividir la cifra obtenida entre la cantidad de valores analizados. Es de anotar que dichos gráficos cuando tienen datos su parte más baja, pero son largos, es considerado este concepto como _"cola"_ ; si la cola se ubica hacia la izquierda es conocida como asmitería negativa, si la cola se ubica  hacia la derecha, se denomina asimetría positiva.

**Ilustración No. 1** - Medidas de Distribución estadística

_Fuente_. Extraído de Romero & Zúnica (2020)

**Ilustración No. 2** - Tipos de simetría

_Fuente_. Extraído de https://estadisticadescriptiva.com/datos/medidas-de-asimetria/

```python
#crear el gráfico de distribución utilizando sns.distplot() 
sns.distplot(df1['Tiempo motor ON (h)']) 
sns.distplot(df1['Tiempo en Movimiento (h)']) 
 
#calculo la media de los datos 
media = df1['Tiempo motor ON (h)'].mean() 
media = df1['Tiempo en Movimiento (h)'].mean() 

plt.title('Gráfico No. 10 - Histogramas de Tiempo motor ON y Tiempo en Movimiento con curvas de asimetría')
plt.xlabel("Tiempo en Movimiento (Horas)")
plt.ylabel('Frecuencia')
plt.show()
```

* Graficos visualizando línea de la media con Matplotlib.

```python
plt.figure(figsize=(10, 6))# configuración del tamaño del gráfico


plt.hist(df1['Tiempo motor ON (h)'], 
         bins=20, color='blue', 
         alpha=0.7, 
         label='Tiempo motor ON (h)')#para crear grafico de histograma de la columna Tiempo motor ON


media_tiempo_motor = df1['Tiempo motor ON (h)'].mean()#media de Tiempo motor ON

plt.axvline(x=media_tiempo_motor, 
            color='red', 
            linestyle='--', 
            label=f'Media: {media_tiempo_motor:.2f} h')# crear la linea de la media


plt.hist(df1['Tiempo en Movimiento (h)'], 
         bins=20, 
         color='orange', 
         alpha=0.7, label='Tiempo en Movimiento (h)')# para crear la grafica de la columna Tiempo en movimiento


media_tiempo_movimiento = df1['Tiempo en Movimiento (h)'].mean()#media de Tiempo en Movimiento (h)

plt.axvline(x=media_tiempo_movimiento, 
            color='red', 
            linestyle='--', 
            label=f'Media: {media_tiempo_movimiento:.2f} h')# crear la linea de la media


plt.xlabel('Tiempo (Horas)')
plt.ylabel('Frecuencia')
plt.title('Gráfico No. 11 - Histogramas Tiempo motor ON y Tiempo en Movimiento (con lineas de media)')
plt.legend()#se añade leyenda añ gráfico para que mueste el dato de la media."
plt.show()
```
